```python
import math

# Action space
# -30, 2
# -20, 2.5
# -15, 2.9
# -10, 3.3
# 0, 3.8
# 0, 3.4
# 10, 3.3
# 15, 2.9
# 20, 2.5
# 20, 2.2
# 30, 1.2

# discount factor = 0.993
# batch size = 256
# Num episodes = 40
# learning rate = 0.0002

optimal_racing_line = [[1,3.06343,0.69254],
[2,3.21149,0.68825],
[3,3.36021,0.68563],
[4,3.50939,0.68424],
[5,3.65888,0.68365],
[6,3.80856,0.68352],
[7,3.95824,0.68375],
[8,4.10771,0.68449],
[9,4.25689,0.6859],
[10,4.4057,0.68814],
[11,4.55405,0.6914],
[12,4.70183,0.69587],
[13,4.84895,0.70175],
[14,4.99528,0.70925],
[15,5.14073,0.71853],
[16,5.28517,0.7298],
[17,5.42847,0.74328],
[18,5.57044,0.75922],
[19,5.71088,0.77798],
[20,5.84943,0.80004],
[21,5.98566,0.82606],
[22,6.11902,0.85672],
[23,6.24883,0.89285],
[24,6.37417,0.93543],
[25,6.49405,0.98532],
[26,6.60668,1.04408],
[27,6.70981,1.11314],
[28,6.80029,1.19386],
[29,6.87318,1.28712],
[30,6.93299,1.3874],
[31,6.97865,1.49365],
[32,7.00767,1.6049],
[33,7.01563,1.71907],
[34,7.00571,1.83281],
[35,6.9796,1.94406],
[36,6.93866,2.05135],
[37,6.88258,2.15283],
[38,6.8089,2.24481],
[39,6.72222,2.32771],
[40,6.62408,2.40119],
[41,6.51674,2.46604],
[42,6.40193,2.52318],
[43,6.2813,2.57388],
[44,6.15613,2.6193],
[45,6.02762,2.6607],
[46,5.89687,2.69935],
[47,5.76502,2.73663],
[48,5.63036,2.77635],
[49,5.49632,2.81779],
[50,5.36302,2.86114],
[51,5.2306,2.90668],
[52,5.09923,2.95476],
[53,4.96914,3.0058],
[54,4.84057,3.06028],
[55,4.71393,3.119],
[56,4.58986,3.18317],
[57,4.46788,3.25171],
[58,4.34771,3.32389],
[59,4.22905,3.39902],
[60,4.11173,3.47661],
[61,3.99556,3.5562],
[62,3.88037,3.63733],
[63,3.76593,3.71958],
[64,3.65242,3.80002],
[65,3.53824,3.87896],
[66,3.42299,3.95547],
[67,3.30623,4.0285],
[68,3.18746,4.09677],
[69,3.06616,4.15877],
[70,2.94184,4.21265],
[71,2.81422,4.25606],
[72,2.68447,4.29081],
[73,2.55319,4.31739],
[74,2.4209,4.33601],
[75,2.28807,4.34686],
[76,2.15517,4.3501],
[77,2.0227,4.34571],
[78,1.89122,4.33367],
[79,1.76134,4.3139],
[80,1.634,4.28541],
[81,1.51044,4.24729],
[82,1.39243,4.19825],
[83,1.28248,4.13677],
[84,1.18432,4.06093],
[85,1.10112,3.97063],
[86,1.03309,3.86834],
[87,0.97851,3.75693],
[88,0.93602,3.63827],
[89,0.90443,3.5137],
[90,0.883,3.38413],
[91,0.87092,3.2504],
[92,0.8672,3.11332],
[93,0.871,2.97352],
[94,0.8814,2.83164],
[95,0.89802,2.68813],
[96,0.92061,2.54372],
[97,0.94879,2.39965],
[98,0.9822,2.2572],
[99,1.02052,2.11724],
[100,1.06366,1.98038],
[101,1.11156,1.84703],
[102,1.16425,1.71754],
[103,1.22184,1.59223],
[104,1.28498,1.47189],
[105,1.35434,1.35741],
[106,1.43079,1.24995],
[107,1.51542,1.151],
[108,1.60967,1.06274],
[109,1.71562,0.98875],
[110,1.82996,0.92599],
[111,1.95115,0.8731],
[112,2.07808,0.82902],
[113,2.20978,0.7927],
[114,2.34551,0.76332],
[115,2.48461,0.74008],
[116,2.62647,0.72219],
[117,2.77054,0.70881],
[118,2.91632,0.69916]]

hair_pin_turn_hard = list(range(20,40)) + list(range(78,88))
hair_pin_turn_soft = [76,77,88,89] + list(range(102,108))
sharp_left_turn = [17,18,19,40,90,91] + list(range(67,76)) + list(range(98,102)) + list(range(108,113))
left_turn_15 = [16, 41, 42, 92, 113] + list(range(62,67))
left_turn_10 = [14, 15, 114] + list(range(93,98))
right_turn = list(range(46,60))

high_speed = list(range(1,15)) + list(range(114,118))
mid_speed = list(range(43,65)) + list(range(92,98))

WAY_POINT_BUFFER = 10
TRACK_WIDTH = 0.76
MIN_REWARD = 0.001
TOT_WAY_POINTS = 118
TARGET_RADIUS = TRACK_WIDTH * 0.9
WORST_DIFF = 15
WORST_DIFF_HAIRPIN = 30
WORST_DIST = TRACK_WIDTH * 0.3
TARGET_STEPS = 130

def dist(point1, point2):
    return ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5

def heron_formula(a, b, c):
    try:
        distance = abs(-(a**4) + 2*(a**2)*(b**2) + 2*(a**2)*(c**2) + 2*(b**2)*(c**2) - (b**4) - (c**4))**0.5 / (2*a)
    except:
        distance = min(b, c)
    return distance

def opt_wps_by_center_wp(closest_waypoint):
    low = int((closest_waypoint - WAY_POINT_BUFFER + TOT_WAY_POINTS) % TOT_WAY_POINTS)
    high = int((closest_waypoint + WAY_POINT_BUFFER) % TOT_WAY_POINTS)
    if low < high:
        return optimal_racing_line[low:high]
    return optimal_racing_line[low:TOT_WAY_POINTS] + optimal_racing_line[0:high]

def get_closest_waypoints_opt_rl(closest_waypoint, x, y):
    opt_wps = opt_wps_by_center_wp(closest_waypoint)
    car = [x, y]
    distances = [dist([p[1], p[2]], car) for p in opt_wps]
    min_dist = min(distances)
    i_closest = distances.index(min_dist)
    
    distances_no_closest = distances.copy()
    distances_no_closest[i_closest] = 999
    sec_min_dist = min(distances_no_closest)
    sec_closest_index = distances_no_closest.index(sec_min_dist)
    
    return opt_wps[i_closest], opt_wps[sec_closest_index]
    
def get_heading_waypoint_opt_rl(closest_waypoint, x, y):
    opt_wps = opt_wps_by_center_wp(closest_waypoint)
    car = [x, y]
    is_inside = [dist([p[1], p[2]], car) < TARGET_RADIUS for p in opt_wps]
    if True in is_inside:
        last_inside_idx = len(is_inside) - 1 - is_inside[::-1].index(True)
        first_outside_idx = last_inside_idx + 1
        if first_outside_idx < len(opt_wps):
            return opt_wps[first_outside_idx]
    return None

def direction_to_chg(from_heading, to_heading):
    chg = to_heading - from_heading
    if chg > 180:
        chg -= 360
    elif chg < -180:
        chg += 360
    return chg

def heading_to_opt_rl_direction_steering(car_heading, target_waypoint_opt_rl, x, y):
    if target_waypoint_opt_rl is None:
        return 999
    target_direction = math.degrees(math.atan2(target_waypoint_opt_rl[2] - y, target_waypoint_opt_rl[1] - x))
    return direction_to_chg(car_heading, target_direction)

def dist_to_opt_rl(closest_waypoint, x, y):
    wp1, wp2 = get_closest_waypoints_opt_rl(closest_waypoint, x, y)
    car = [x, y]
    wp1_coord = [wp1[1], wp1[2]]
    wp2_coord = [wp2[1], wp2[2]]
    
    car_to_wp1 = abs(dist(car, wp1_coord))
    car_to_wp2 = abs(dist(car, wp2_coord))
    wp_dist = abs(dist(wp1_coord, wp2_coord))
    return heron_formula(wp_dist, car_to_wp1, car_to_wp2)
    
def reward_by_diff(weight, diff, worst_diff):
    error = diff / worst_diff
    score = weight * (1 - abs(error))
    return max(score, MIN_REWARD)
    
def reward_distance(distance):
    return reward_by_diff(1.0, distance, WORST_DIST)
    
def reward_direction(direction_diff, steering_angle, closest_waypoint, speed):
    diff = abs(direction_diff - steering_angle)
    weight = 1.0
    worst_diff = WORST_DIFF
    # hair-pin turn reward
    if closest_waypoint in hair_pin_turn_hard:
        worst_diff = WORST_DIFF_HAIRPIN
        if steering_angle == 30:
            weight = 10.0
        else:
            weight = MIN_REWARD
    elif closest_waypoint in hair_pin_turn_soft:
        worst_diff = WORST_DIFF_HAIRPIN
        if steering_angle == 30:
            weight = 10.0
        elif steering_angle <= 15:
            weight = MIN_REWARD
    # should left turn sharply
    elif closest_waypoint in sharp_left_turn:
        if steering_angle >= 20:
            weight = 5.0
        elif steering_angle <= 10:
            weight = MIN_REWARD
    # should left turn for 10'
    elif closest_waypoint in left_turn_10:
        if steering_angle >= 10:
            weight = 3.0
        elif steering_angle < 0:
            weight = MIN_REWARD
    # should left turn for 15'
    elif closest_waypoint in left_turn_15:
        if steering_angle >= 15:
            weight = 3.0
        elif steering_angle <= 0:
            weight = MIN_REWARD
    # should right turn
    elif closest_waypoint in right_turn:
        if steering_angle < 0:
            weight = 3.0
        elif steering_angle >= 15:
            weight = MIN_REWARD
            
    # high speed
    if closest_waypoint in high_speed:
        if speed > 3:
            weight *= 3.0
        elif speed <= 2.5:
            weight *= 0.1
            
    elif closest_waypoint in mid_speed:
        if speed < 3.5 and speed > 2.5:
            weight *= 3
        elif speed <= 2.2:
            weight *= 0.1
            
    return reward_by_diff(weight, diff, worst_diff)

def reward_function(params):
    x = params['x']
    y = params['y']
    heading = params['heading']
    steering_angle = params['steering_angle']
    closest_waypoints = params['closest_waypoints']
    speed = params['speed']
    steps = params['steps']
    progress = params['progress']    
    closest_waypoint = closest_waypoints[0]
    
    target_waypoint_opt_rl = get_heading_waypoint_opt_rl(closest_waypoint, x, y)
    
    if target_waypoint_opt_rl is None:
        return MIN_REWARD
    
    direction_diff = heading_to_opt_rl_direction_steering(heading, target_waypoint_opt_rl, x, y)
    direction_reward = float(reward_direction(direction_diff, steering_angle, closest_waypoint, speed))

    distance = dist_to_opt_rl(closest_waypoint, x, y)
    dist_reward = float(reward_distance(distance))
    
    if direction_reward <= MIN_REWARD:
        return MIN_REWARD
    
    step_reward = ((direction_reward + dist_reward) ** 1.3 + dist_reward * direction_reward) ** 0.75
    
    return float(step_reward)
```
