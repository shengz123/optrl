```python
import math

# Action space
# -30, 1.6
# -10, 2.7
# 0, 4
# 0, 3.6
# 0, 3.4
# 10, 2.8
# 15, 2.6
# 20, 2.4
# 30, 1.2

optimal_racing_line = [[1,3.05991,0.68275,15,0.5],
[2,3.20949,0.68254,15,0.5],
[3,3.35926,0.68295,15,0.5],
[4,3.50903,0.68327,15,0.5],
[5,3.65879,0.68346,15,0.5],
[6,3.80856,0.68352,15,0.5],
[7,3.95832,0.68357,15,0.5],
[8,4.10808,0.68362,15,0.5],
[9,4.25783,0.68367,15,0.5],
[10,4.40753,0.68392,15,0.5],
[11,4.55708,0.68458,15,0.5],
[12,4.7064,0.68591,15,0.5],
[13,4.85538,0.68818,15,0.5],
[14,5.00391,0.69167,15,0.5],
[15,5.15186,0.6967,15,0.5],
[16,5.29909,0.70359,15,0.5],
[17,5.44542,0.71271,15,0.5],
[18,5.59064,0.72448,15,0.5],
[19,5.7345,0.73937,15,0.8],
[20,5.87671,0.75794,15,0.8],
[21,6.01688,0.78079,15,0.8],
[22,6.15453,0.80868,15,0.8],
[23,6.28892,0.84268,15,0.8],
[24,6.41915,0.88391,15,0.8],
[25,6.54402,0.93369,15,0.8],
[26,6.66177,0.99362,15,0.8],
[27,6.77031,1.06511,15,0.8],
[28,6.86577,1.15033,15,0.8],
[29,6.94226,1.25049,30,0.8],
[30,7.00517,1.35858,30,0.8],
[31,7.05301,1.47356,40,0.8],
[32,7.08346,1.59413,40,0.8],
[33,7.09215,1.71798,50,0.8],
[34,7.08146,1.84137,50,0.8],
[35,7.05368,1.96193,50,0.8],
[36,7.01029,2.07799,40,0.8],
[37,6.95148,2.18767,40,0.8],
[38,6.87449,2.28676,40,0.8],
[39,6.78416,2.37562,40,0.5],
[40,6.68126,2.45276,40,0.5],
[41,6.56787,2.5182,30,0.5],
[42,6.44654,2.57349,30,0.5],
[43,6.31897,2.61994,30,0.5],
[44,6.18655,2.65895,15,0.5],
[45,6.0506,2.6922,15,0.5],
[46,5.91225,2.72138,15,0.5],
[47,5.77246,2.74803,15,0.5],
[48,5.63119,2.77387,15,0.5],
[49,5.49054,2.80146,15,0.5],
[50,5.35082,2.83151,15,0.5],
[51,5.21237,2.86477,15,0.5],
[52,5.07557,2.90203,15,0.5],
[53,4.94098,2.94442,15,0.5],
[54,4.80919,2.99314,15,0.5],
[55,4.68108,3.04995,15,0.5],
[56,4.55772,3.11677,15,0.5],
[57,4.43809,3.19111,15,0.5],
[58,4.32158,3.27156,15,0.5],
[59,4.20763,3.3569,15,0.5],
[60,4.09584,3.44634,15,0.5],
[61,3.98587,3.53911,15,0.5],
[62,3.87735,3.63453,15,0.5],
[63,3.77153,3.73097,15,0.5],
[64,3.66425,3.82459,15,0.5],
[65,3.55522,3.91482,15,0.8],
[66,3.44413,4.00101,15,0.8],
[67,3.33062,4.08243,15,0.8],
[68,3.21408,4.1578,15,0.8],
[69,3.0938,4.22557,15,0.8],
[70,2.96902,4.28399,15,0.8],
[71,2.83884,4.33041,15,0.8],
[72,2.70477,4.36698,15,0.8],
[73,2.56773,4.39457,15,0.8],
[74,2.42858,4.41372,15,0.8],
[75,2.28804,4.42463,15,0.8],
[76,2.14689,4.42745,15,0.8],
[77,2.00594,4.42207,15,0.8],
[78,1.86608,4.40842,15,0.8],
[79,1.72824,4.3862,15,0.8],
[80,1.59367,4.35463,15,0.8],
[81,1.46401,4.3126,15,0.8],
[82,1.34122,4.25902,15,0.8],
[83,1.22772,4.19264,15,0.8],
[84,1.12683,4.11194,15,0.8],
[85,1.04082,4.01742,15,0.8],
[86,0.97007,3.91116,15,0.8],
[87,0.91616,3.79418,30,0.8],
[88,0.87678,3.66946,30,0.5],
[89,0.8504,3.53881,30,0.5],
[90,0.8355,3.40361,45,0.5],
[91,0.83064,3.26495,45,0.5],
[92,0.83445,3.12371,45,0.5],
[93,0.84572,2.98056,30,0.5],
[94,0.86309,2.83609,30,0.5],
[95,0.88531,2.69072,15,0.5],
[96,0.91122,2.5448,15,0.5],
[97,0.94016,2.3985,15,0.5],
[98,0.97149,2.25202,15,0.5],
[99,1.00561,2.10736,15,0.5],
[100,1.04318,1.96498,15,0.5],
[101,1.08485,1.82544,15,0.8],
[102,1.13138,1.68946,15,0.8],
[103,1.18354,1.55779,15,0.8],
[104,1.2421,1.43127,15,0.8],
[105,1.30792,1.31094,15,0.8],
[106,1.38211,1.19819,15,0.8],
[107,1.46587,1.09472,15,0.8],
[108,1.56114,1.00344,15,0.8],
[109,1.6701,0.92861,15,0.5],
[110,1.78853,0.8668,30,0.5],
[111,1.91462,0.81665,30,0.5],
[112,2.04674,0.77661,15,0.5],
[113,2.18374,0.74556,15,0.5],
[114,2.32461,0.72235,15,0.5],
[115,2.46847,0.70582,15,0.5],
[116,2.61453,0.69478,15,0.5],
[117,2.76212,0.688,15,0.5],
[118,2.9107,0.68431,15,0.5]]

sharp_turn_targets = list(range(29, 44))

WAY_POINT_BUFFER = 10
TRACK_WIDTH = 0.76
MIN_REWARD = 0.001
TOT_WAY_POINTS = 118

def dist(point1, point2):
    return ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5

def opt_wps_by_center_wp(cloest_center_line_wp):
    low = int((cloest_center_line_wp - WAY_POINT_BUFFER + TOT_WAY_POINTS) % TOT_WAY_POINTS)
    high = int((cloest_center_line_wp + WAY_POINT_BUFFER) % TOT_WAY_POINTS)
    if low < high:
        return optimal_racing_line[low:high]
    return optimal_racing_line[low:TOT_WAY_POINTS] + optimal_racing_line[0:high]

def get_heading_waypoint_opt_rl(cloest_center_line_wp, x, y):
    opt_wps = opt_wps_by_center_wp(cloest_center_line_wp)
    radius = TRACK_WIDTH * optimal_racing_line[cloest_center_line_wp-1][4]
    car = [x, y]
    is_inside = [dist([p[1], p[2]], car) < radius for p in opt_wps]
    if True in is_inside:
        last_inside_idx = len(is_inside) - 1 - is_inside[::-1].index(True)
        first_outside_idx = last_inside_idx + 1
        if first_outside_idx < len(opt_wps):
            return opt_wps[first_outside_idx]
    return None

def direction_to_chg(from_heading, to_heading):
    chg = to_heading - from_heading
    if chg > 180:
        chg -= 360
    elif chg < -180:
        chg += 360
    return chg

def heading_to_opt_rl_direction_steering(car_heading, target_waypoint_opt_rl, x, y):
    if target_waypoint_opt_rl is None:
        return 999
    target_direction = math.degrees(math.atan2(target_waypoint_opt_rl[2] - y, target_waypoint_opt_rl[1] - x))
    return direction_to_chg(car_heading, target_direction)

def reward_by_diff(weight, diff, worst_diff):
    error = diff / worst_diff
    score = weight * (1 - abs(error))
    return max(score, MIN_REWARD)

def reward_direction(direction_diff, steering_angle, worst_diff, target_way_point):
    diff = abs(direction_diff - steering_angle)
    weight = 1.0
    if target_way_point in sharp_turn_targets and steering_angle == 30:
        weight = 3.0
    return reward_by_diff(weight, diff, worst_diff)

def reward_function(params):
    x = params['x']
    y = params['y']
    heading = params['heading']
    steering_angle = params['steering_angle']
    closest_waypoints = params['closest_waypoints']
    closest_waypoint = closest_waypoints[0]
    target_waypoint_opt_rl = get_heading_waypoint_opt_rl(closest_waypoint, x, y)
    if target_waypoint_opt_rl is None:
        return MIN_REWARD
    direction_diff = heading_to_opt_rl_direction_steering(heading, target_waypoint_opt_rl, x, y)
    return float(reward_direction(direction_diff, steering_angle, target_waypoint_opt_rl[3], target_waypoint_opt_rl[0]))
```
