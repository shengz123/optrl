```python
import math

# Action space
# -30, 2
# -20, 2.5
# -15, 2.9
# -10, 3.3
# 0, 3.8
# 0, 3.4
# 10, 3.2
# 15, 2.9
# 20, 2.5
# 20, 2.3
# 30, 1.7
# 30, 1.2

# discount factor = 0.993
# batch size = 256
# Num episodes = 40
# learning rate = 0.0002
# Training hours: 6 hours

optimal_racing_line = [[1,3.05991,0.68275],
[2,3.20949,0.68254],
[3,3.35926,0.68295],
[4,3.50903,0.68327],
[5,3.65879,0.68346],
[6,3.80856,0.68352],
[7,3.95832,0.68357],
[8,4.10808,0.68362],
[9,4.25783,0.68367],
[10,4.40753,0.68392],
[11,4.55708,0.68458],
[12,4.7064,0.68591],
[13,4.85538,0.68818],
[14,5.00391,0.69167],
[15,5.15186,0.6967],
[16,5.29909,0.70359],
[17,5.44542,0.71271],
[18,5.59064,0.72448],
[19,5.7345,0.73937],
[20,5.87671,0.75794],
[21,6.01688,0.78079],
[22,6.15453,0.80868],
[23,6.28892,0.84268],
[24,6.41915,0.88391],
[25,6.54402,0.93369],
[26,6.66177,0.99362],
[27,6.77031,1.06511],
[28,6.86577,1.15033],
[29,6.94226,1.25049],
[30,7.00517,1.35858],
[31,7.05301,1.47356],
[32,7.08346,1.59413],
[33,7.09215,1.71798],
[34,7.08146,1.84137],
[35,7.05368,1.96193],
[36,7.01029,2.07799],
[37,6.95148,2.18767],
[38,6.87449,2.28676],
[39,6.78416,2.37562],
[40,6.68126,2.45276],
[41,6.56787,2.5182],
[42,6.44654,2.57349],
[43,6.31897,2.61994],
[44,6.18655,2.65895],
[45,6.0506,2.6922],
[46,5.91225,2.72138],
[47,5.77246,2.74803],
[48,5.63119,2.77387],
[49,5.49054,2.80146],
[50,5.35082,2.83151],
[51,5.21237,2.86477],
[52,5.07557,2.90203],
[53,4.94098,2.94442],
[54,4.80919,2.99314],
[55,4.68108,3.04995],
[56,4.55772,3.11677],
[57,4.43809,3.19111],
[58,4.32158,3.27156],
[59,4.20763,3.3569],
[60,4.09584,3.44634],
[61,3.98587,3.53911],
[62,3.87735,3.63453],
[63,3.77153,3.73097],
[64,3.66425,3.82459],
[65,3.55522,3.91482],
[66,3.44413,4.00101],
[67,3.33062,4.08243],
[68,3.21408,4.1578],
[69,3.0938,4.22557],
[70,2.96902,4.28399],
[71,2.83884,4.33041],
[72,2.70477,4.36698],
[73,2.56773,4.39457],
[74,2.42858,4.41372],
[75,2.28804,4.42463],
[76,2.14689,4.42745],
[77,2.00594,4.42207],
[78,1.86608,4.40842],
[79,1.72824,4.3862],
[80,1.59367,4.35463],
[81,1.46401,4.3126],
[82,1.34122,4.25902],
[83,1.22772,4.19264],
[84,1.12683,4.11194],
[85,1.04082,4.01742],
[86,0.97007,3.91116],
[87,0.91616,3.79418],
[88,0.87678,3.66946],
[89,0.8504,3.53881],
[90,0.8355,3.40361],
[91,0.83064,3.26495],
[92,0.83445,3.12371],
[93,0.84572,2.98056],
[94,0.86309,2.83609],
[95,0.88531,2.69072],
[96,0.91122,2.5448],
[97,0.94016,2.3985],
[98,0.97149,2.25202],
[99,1.00561,2.10736],
[100,1.04318,1.96498],
[101,1.08485,1.82544],
[102,1.13138,1.68946],
[103,1.18354,1.55779],
[104,1.2421,1.43127],
[105,1.30792,1.31094],
[106,1.38211,1.19819],
[107,1.46587,1.09472],
[108,1.56114,1.00344],
[109,1.6701,0.92861],
[110,1.78853,0.8668],
[111,1.91462,0.81665],
[112,2.04674,0.77661],
[113,2.18374,0.74556],
[114,2.32461,0.72235],
[115,2.46847,0.70582],
[116,2.61453,0.69478],
[117,2.76212,0.688],
[118,2.9107,0.68431]]

# if steering_angle = 30 and speed = 1.2, reward *= 10; if steering_angle <= 15, reward *= 0.1
hair_pin_turn = list(range(27, 46))

# if steering_angle = 30 and speed = 1.7, reward *= 10; if steering_angle <= 15, reward *= 0.1
hair_pin_turn_mid = list(range(84, 96))

# if steering_angle >= 10 and reward *= 3; if steering_angle < 0, reward *= 0.1 
left_turn = list(range(20, 27)) + list(range(46, 52)) + list(range(96, 111))

# if steering_angle >= 15, reward *= 5; if steering_angle <= 10, reward *= 0.1
sharp_left_turn = [1] + list(range(72, 84)) + list(range(111,119))

# if steering angle < 0, reward *= 3; if steering_angle >= 15, reward *= 0.1 
right_steering_line = list(range(52,69))

# if speed > 3, reward *=3; if speed <= 2.5, reward *= 0.3
fast_line = list(range(2,21))

# if 3.5 > speed > 2.5, reward *= 1.5; if speed <= 2.2, reward *= 0.3
mid_line = [21,22] + list(range(50, 72)) + list(range(96, 108)) + [117, 118]
mid_line_chk_steering = list(range(96, 108))

start_wps_no_reward = list(range(20, 50))
reward_past_hairpin = list(range(46, 58))

WAY_POINT_BUFFER = 10
TRACK_WIDTH = 0.76
MIN_REWARD = 0.001
TOT_WAY_POINTS = 118
TARGET_RADIUS = TRACK_WIDTH * 0.9
WORST_DIFF = 15
WORST_DIFF_HAIRPIN = 30
WORST_DIST = TRACK_WIDTH * 0.3
TARGET_STEPS = 130

class EpisodeCtx:
    prev_steps = None
    starting_wp = None
    
def set_episode(steps, waypoint):
    if EpisodeCtx.prev_steps is None or steps < EpisodeCtx.prev_steps:
        EpisodeCtx.starting_wp = waypoint
                
    EpisodeCtx.prev_steps = steps
    
def dist(point1, point2):
    return ((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2) ** 0.5

def heron_formula(a, b, c):
    try:
        distance = abs(-(a**4) + 2*(a**2)*(b**2) + 2*(a**2)*(c**2) + 2*(b**2)*(c**2) - (b**4) - (c**4))**0.5 / (2*a)
    except:
        distance = min(b, c)
    return distance

def opt_wps_by_center_wp(closest_waypoint):
    low = int((closest_waypoint - WAY_POINT_BUFFER + TOT_WAY_POINTS) % TOT_WAY_POINTS)
    high = int((closest_waypoint + WAY_POINT_BUFFER) % TOT_WAY_POINTS)
    if low < high:
        return optimal_racing_line[low:high]
    return optimal_racing_line[low:TOT_WAY_POINTS] + optimal_racing_line[0:high]

def get_closest_waypoints_opt_rl(closest_waypoint, x, y):
    opt_wps = opt_wps_by_center_wp(closest_waypoint)
    car = [x, y]
    distances = [dist([p[1], p[2]], car) for p in opt_wps]
    min_dist = min(distances)
    i_closest = distances.index(min_dist)
    
    distances_no_closest = distances.copy()
    distances_no_closest[i_closest] = 999
    sec_min_dist = min(distances_no_closest)
    sec_closest_index = distances_no_closest.index(sec_min_dist)
    
    return opt_wps[i_closest], opt_wps[sec_closest_index]
    
def get_heading_waypoint_opt_rl(closest_waypoint, x, y):
    opt_wps = opt_wps_by_center_wp(closest_waypoint)
    car = [x, y]
    is_inside = [dist([p[1], p[2]], car) < TARGET_RADIUS for p in opt_wps]
    if True in is_inside:
        last_inside_idx = len(is_inside) - 1 - is_inside[::-1].index(True)
        first_outside_idx = last_inside_idx + 1
        if first_outside_idx < len(opt_wps):
            return opt_wps[first_outside_idx]
    return None

def direction_to_chg(from_heading, to_heading):
    chg = to_heading - from_heading
    if chg > 180:
        chg -= 360
    elif chg < -180:
        chg += 360
    return chg

def heading_to_opt_rl_direction_steering(car_heading, target_waypoint_opt_rl, x, y):
    if target_waypoint_opt_rl is None:
        return 999
    target_direction = math.degrees(math.atan2(target_waypoint_opt_rl[2] - y, target_waypoint_opt_rl[1] - x))
    return direction_to_chg(car_heading, target_direction)

def dist_to_opt_rl(closest_waypoint, x, y):
    wp1, wp2 = get_closest_waypoints_opt_rl(closest_waypoint, x, y)
    car = [x, y]
    wp1_coord = [wp1[1], wp1[2]]
    wp2_coord = [wp2[1], wp2[2]]
    
    car_to_wp1 = abs(dist(car, wp1_coord))
    car_to_wp2 = abs(dist(car, wp2_coord))
    wp_dist = abs(dist(wp1_coord, wp2_coord))
    return heron_formula(wp_dist, car_to_wp1, car_to_wp2)
    
def reward_by_diff(weight, diff, worst_diff):
    error = diff / worst_diff
    score = weight * (1 - abs(error))
    return max(score, MIN_REWARD)
    
def reward_distance(distance):
    return reward_by_diff(1.0, distance, WORST_DIST)
    
def reward_direction(direction_diff, steering_angle, target_way_point, speed):
    diff = abs(direction_diff - steering_angle)
    weight = 1.0
    worst_diff = WORST_DIFF
    # hair-pin turn reward
    if target_way_point in hair_pin_turn:
        worst_diff = WORST_DIFF_HAIRPIN
        if steering_angle == 30 and speed < 1.3:
            weight = 10.0
        elif steering_angle <= 15:
            weight = MIN_REWARD
    elif target_way_point in hair_pin_turn_mid:
        worst_diff = WORST_DIFF_HAIRPIN
        if steering_angle == 30 and speed > 1.3:
            weight = 10.0
        elif steering_angle <= 15:
            weight = MIN_REWARD
    # should left turn
    elif target_way_point in left_turn:
        if steering_angle >= 10:
            weight = 3.0
        elif steering_angle < 0:
            weight = 0.1
    # should left turn sharply
    elif target_way_point in sharp_left_turn:
        if steering_angle >= 15:
            weight = 5.0
        elif steering_angle <= 10:
            weight = 0.1
    # should right turn
    elif target_way_point in right_steering_line:
        if steering_angle <= 0:
            weight = 3.0
        elif steering_angle >= 15:
            weight = 0.1
    # high speed
    if target_way_point in fast_line:
        if speed > 3:
            weight *= 2.0
        elif speed <= 2.5:
            weight *= 0.1
    elif target_way_point in mid_line:
        if speed < 3.5 and speed > 2.5:
            if target_way_point not in mid_line_chk_steering or weight >= 0.2:
                weight *= 1.5
        elif speed <= 2.2:
            weight *= 0.1
            
    if EpisodeCtx.starting_wp not in start_wps_no_reward and target_way_point in reward_past_hairpin:
        weight *= (speed * 4)
        
    return reward_by_diff(weight, diff, worst_diff)

def reward_progress(progress, steps):
    if progress == 100 and steps < TARGET_STEPS and EpisodeCtx.starting_wp not in start_wps_no_reward:
        return 100 * (TARGET_STEPS - steps)
    return MIN_REWARD
        
def reward_function(params):
    x = params['x']
    y = params['y']
    heading = params['heading']
    steering_angle = params['steering_angle']
    closest_waypoints = params['closest_waypoints']
    speed = params['speed']
    steps = params['steps']
    progress = params['progress']    
    closest_waypoint = closest_waypoints[0]
    
    set_episode(steps, closest_waypoint)
    
    target_waypoint_opt_rl = get_heading_waypoint_opt_rl(closest_waypoint, x, y)
    
    if target_waypoint_opt_rl is None:
        return MIN_REWARD
    
    direction_diff = heading_to_opt_rl_direction_steering(heading, target_waypoint_opt_rl, x, y)
    direction_reward = float(reward_direction(direction_diff, steering_angle, target_waypoint_opt_rl[0], speed))

    distance = dist_to_opt_rl(closest_waypoint, x, y)
    dist_reward = float(reward_distance(distance))
    
    if direction_reward <= MIN_REWARD:
        return MIN_REWARD
    
    step_reward = ((direction_reward + dist_reward) ** 1.3 + dist_reward * direction_reward) ** 0.75
    
    if dist_reward > MIN_REWARD:
        step_reward += reward_progress(progress, steps)
        
    return float(step_reward)
```
